## 建立字典

```python
import csv  #載入專門開csv檔案的套件
import jieba
import codecs
import jieba.analyse  #載入能萃取出文章中關鍵詞的功能
jieba.set_dictionary('dict.txt')

factory = []                                    
factories = open('factory.txt','r',encoding='utf8').readlines()
for f in factories:
    factory.append(f.strip())
    
car = []                                    
cars = open('car.txt','r',encoding='utf8').readlines()
for c in cars:
    car.append(c.strip())
    
warehouse = []                                    
warehouses = open('warehouse.txt','r',encoding='utf8').readlines()
for w in warehouses:
    warehouse.append(w.strip())
    
business = []                                    
businesses = open('business.txt','r',encoding='utf8').readlines()
for b in businesses:
    business.append(b.strip())
    
store = []                                    
stores = open('store.txt','r',encoding='utf8').readlines()
for st in stores:
    store.append(st.strip())
    
house = []                                    
houses = open('house.txt','r',encoding='utf8').readlines()
for h in houses:
    house.append(h.strip())
    
    
tao = []                                    
taos = open('tao.txt','r',encoding='utf8').readlines()
for t in taos:
    tao.append(t.strip())
    
ya = []                                    
yas = open('ya.txt','r',encoding='utf8').readlines()
for y in yas:
    ya.append(y.strip())

stop = []                                    
stops = open('stop.txt','r',encoding='utf8').readlines()
for s in stops:
    stop.append(s.strip())
```

```python
content = []   #裝最後分詞完的結果
# contents = open('1.csv','r',encoding='utf-8', errors='ignore').readlines()    #開啟檔案
for row in csv.DictReader(contents):    #csv.DictReader函式可依欄位名開啟特定欄位
    words = jieba.cut(row['title'].replace(' ',''), cut_all=False)    #只讓'name'此欄位下的內容做斷詞
    temp = []    #每個詞斷出的結果
    category = []
    for word in words:
        if word not in stop:
            temp.append(word)   
            

        if word in tao :         
            category.append('t')
            break
        if word in ya :         
            category.append('y')
            break 
        if word in factory :          
            category.append('f')
            break
        if word in warehouse :          
            category.append('w')
            break
        if word in business :       
            category.append('b')
            break
        if word in store :         
            category.append('s')
            break
        if word in house :         
            category.append('h')
            break
        if word in car :         
            category.append('c')
            break

    print(category)
    
    if category:        #如果category不為空字串的話 （空字串在python if條件式下為false）
        content.append(temp+category)
        
    else:
        content.append(temp)
   
    
# content
```
